## Difference between Merging, Joining, and Concatenation
In summary:

*   **Merge**: Combine based on shared columns or indices (like SQL joins).
*   **Join**: Combine based on the index (a special case of merging).
*   **Concatenate**: Stack DataFrames together along an axis.

##GroupBy

```python
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Store': ['S1', 'S1', 'S2', 'S2', 'S1', 'S2', 'S2', 'S1'],
    'Sales': [100, 200, 150, 250, 120, 180, 200, 300],
    'Quantity': [10, 15, 12, 18, 8, 20, 15, 25],
    'Date': pd.date_range('2023-01-01', periods=8)
}
df = pd.DataFrame(data)
```

##**PIVOT_TABLES**

```python
data = {
    'Date': pd.date_range('2023-01-01', periods=20),
    'Product': ['A', 'B', 'C', 'D'] * 5,
    'Region': ['East', 'West', 'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West',
               'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South'],
    'Sales': np.random.randint(100, 1000, 20),
    'Units': np.random.randint(10, 100, 20),
    'Rep': ['John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary',
            'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice']
}

df = pd.DataFrame(data)

df['Month'] = df['Date'].dt.month_name()
df['Quarter'] = 'Q' + df['Date'].dt.quarter.astype(str)
df
```

```python
pd.pivot_table(df,values = "Sales",index = 'Region',columns="Product",aggfunc = 'median')
```

```python
pivot2 = pd.pivot_table(df, values=['Sales', 'Units'], index='Region', columns='Product')
pivot2
```

##Cross Tabs

```python
pd.crosstab(df['Region'],df['Product'])
```

## Difference between Pivot Tables and Crosstabs

Both pivot tables and crosstabs are powerful tools in pandas for summarizing data in a DataFrame, but they are used in slightly different ways:

*   **Pivot Table**: A pivot table is used to reshape data and provide a summary based on one or more keys. It's more flexible than a crosstab and can perform various aggregation functions (like sum, mean, count, etc.) on specified values. You define the index, columns, and values you want to aggregate, and the function you want to use for aggregation.

*   **Crosstab**: A crosstab (or cross-tabulation) is a special case of a pivot table that is specifically used to compute a frequency table of two or more factors. By default, it calculates the count of occurrences for each combination of the specified rows and columns. It's primarily used for categorical data to understand the relationship between different categories.
