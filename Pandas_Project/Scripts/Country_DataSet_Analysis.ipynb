## 1Ô∏è‚É£ Which country has the highest population?

## 2Ô∏è‚É£ What is the capital of the country with highest population?


```python
df[df['population'] == df['population'].max()]['capital_city']
```

## 3Ô∏è‚É£ Which country has the least population?


```python
df[df['population'] == df['population'].min()]['country']
```

## 4Ô∏è‚É£ What is the capital of the country with least population?

## 5Ô∏è‚É£ Give me Top 5 countries with highest democratic score


```python
df.sort_values(by = 'democracy_score',ascending=False,inplace = True)
```

## 6Ô∏è‚É£ How many total regions are there?


```python
df['region'].value_counts().count()
```

## 7Ô∏è‚É£ How many countries lie in Eastern Europe region?


```python
df[df['region'] == "Eastern Europe"][ 'country']
```

## 8Ô∏è‚É£ Who is the political leader of the 2nd highest populated country?


```python
df[df['population'] == df['population'].nlargest(2).iloc[1]]['political_leader']
```

## 9Ô∏è‚É£ How many countries are there whose political leaders are unknown?


```python
df[df['political_leader'].isna()]['country'].count()
```

## üîü How many countries have Republic in their full name?


```python
count = 0
def counting(txt):
    global count
    if 'republic' in txt.lower():
        count+= 1
    return txt
df['country_long'] = df['country_long'].apply(counting)
print(count)
```
